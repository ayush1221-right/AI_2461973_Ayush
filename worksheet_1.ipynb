```python
import numpy as np

# Define the matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Given matrices:")
print(f"A = \n{A}")
print(f"\nB = \n{B}")
```

```python
# Task 1: Prove A.A^(-1) = I
print("\n" + "="*60)
print("1. Prove A·A^(-1) = I")
print("="*60)
A_inv = np.linalg.inv(A)
print(f"A^(-1) = \n{A_inv}")

product_A_Ainv = np.dot(A, A_inv)
print(f"\nA·A^(-1) = \n{product_A_Ainv}")

# Round to remove floating point errors
product_rounded = np.round(product_A_Ainv)
print(f"\nA·A^(-1) (rounded) = \n{product_rounded}")

identity_2x2 = np.eye(2)
print(f"\nIdentity matrix I = \n{identity_2x2}")
print(f"\nVerification: A·A^(-1) = I? {np.allclose(product_A_Ainv, identity_2x2)}")
```

```python
# Task 2: Prove AB ≠ BA
print("\n" + "="*60)
print("2. Prove AB ≠ BA")
print("="*60)
AB = np.dot(A, B)
BA = np.dot(B, A)

print(f"AB = \n{AB}")
print(f"\nBA = \n{BA}")
print(f"\nAre AB and BA equal? {np.array_equal(AB, BA)}")
print("Therefore, AB ≠ BA (Matrix multiplication is not commutative)")
```

```python
# Task 3: Prove (AB)^T = B^T·A^T
print("\n" + "="*60)
print("3. Prove (AB)^T = B^T·A^T")
print("="*60)
AB_transpose = AB.T
print(f"(AB)^T = \n{AB_transpose}")

B_transpose = B.T
A_transpose = A.T
print(f"\nB^T = \n{B_transpose}")
print(f"\nA^T = \n{A_transpose}")

BT_AT = np.dot(B_transpose, A_transpose)
print(f"\nB^T·A^T = \n{BT_AT}")

print(f"\nAre (AB)^T and B^T·A^T equal? {np.array_equal(AB_transpose, BT_AT)}")
print("Therefore, (AB)^T = B^T·A^T ✓")
```

```python
 #Task 4: Solve system of linear equations
print("\n" + "="*60)
print("4. Solve System of Linear Equations using Inverse Method")
print("="*60)
print("System of equations:")
print("2x - 3y + z = -1")
print("x - y + 2z = -3")
print("3x + y - z = 9")

# Coefficient matrix A
coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])

# Constants matrix B
constants = np.array([-1, -3, 9])

print(f"\nCoefficient matrix A = \n{coeff_matrix}")
print(f"\nConstants matrix B = \n{constants}")

#  Using matrix inverse
print("\n--- Using Matrix Inverse (A^(-1)·B) ---")
coeff_inv = np.linalg.inv(coeff_matrix)
print(f"A^(-1) = \n{coeff_inv}")

solution_method1 = np.dot(coeff_inv, constants)
print(f"\nSolution X = A^(-1)·B = \n{solution_method1}")
print(f"\nTherefore:")
print(f"x = {solution_method1[0]}")
print(f"y = {solution_method1[1]}")
print(f"z = {solution_method1[2]}")
```

```python
# Verification
print("\n--- Verification: Check if A·X = B ---")
verification = np.dot(coeff_matrix, solution_method1)
print(f"A·X = \n{verification}")
print(f"B = \n{constants}")
print(f"Is A·X = B? {np.allclose(verification, constants)}")

print("\nAll tasks completed!")
```
