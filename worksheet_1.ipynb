```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('Titanic-Dataset.csv')

# Create a DataFrame subsetted for specific columns
df_subset = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]

# Retain only rows where Pclass is equal to 1 (first-class passengers)
df_first_class = df_subset[df_subset['Pclass'] == 1]

# Calculate statistics for the Fare column
mean_fare = df_first_class['Fare'].mean()
median_fare = df_first_class['Fare'].median()
max_fare = df_first_class['Fare'].max()
min_fare = df_first_class['Fare'].min()

print("Statistics for First-Class Passengers' Fare:")
print(f"Mean: {mean_fare}")
print(f"Median: {median_fare}")
print(f"Maximum: {max_fare}")
print(f"Minimum: {min_fare}")
```

```python
# Problem 2: Check null values in Age column
print("\nNull values in Age column:")
null_count = df_first_class['Age'].isnull().sum()
print(f"Number of null values: {null_count}")
```

```python
# Drop null values from Age column
df_first_class = df_first_class.dropna(subset=['Age'])
```

```python
 #Problem 3: One-hot encoding for Embarked column
# Load the full dataset again to access Embarked column
df_full = pd.read_csv('Titanic-Dataset.csv')
```

```python
# Use one-hot encoding to convert Embarked column
embarked_encoded = pd.get_dummies(df_full['Embarked'], prefix='Embarked')
```

```python
# Add these new columns to the original DataFrame
df_full = pd.concat([df_full, embarked_encoded], axis=1)

print("\nOne-hot encoded Embarked columns added to original DataFrame:")
print(df_full[['Embarked', 'Embarked_C', 'Embarked_Q', 'Embarked_S']].head())
print(f"\nDataFrame shape after adding encoded columns: {df_full.shape}")
```

```python
# Drop the original Embarked column
df_full = df_full.drop(columns=['Embarked'])

print("\nAfter dropping original Embarked column:")
print(df_full[['Embarked_C', 'Embarked_Q', 'Embarked_S']].head())
print(f"DataFrame shape after dropping Embarked: {df_full.shape}")
```

```python
# Problem 4: Compare mean survival rates by Sex
import matplotlib.pyplot as plt

# Load original dataset for this analysis
df_original = pd.read_csv('Titanic-Dataset.csv')

# Calculate mean survival rates by Sex
survival_by_sex = df_original.groupby('Sex')['Survived'].mean()

print("\nMean Survival Rates by Sex:")
print(survival_by_sex)

# Create visualization
plt.figure()
survival_by_sex.plot(kind='bar', color=['skyblue', 'pink'], title='Survival Rate by Gender')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

```python
# Problem 5: Survival rates by Sex and Embarked port
# Calculate mean survival rates by Sex and Embarked
survival_by_sex_embarked = df_original.groupby(['Sex', 'Embarked'])['Survived'].mean().unstack()

print("\nMean Survival Rates by Sex and Embarked Port:")
print(survival_by_sex_embarked)

# Create visualization
plt.figure()
survival_by_sex_embarked.plot(kind='bar', title='Survival Rate by Gender and Port of Embarkation')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.legend(title='Embarked', labels=['Cherbourg (C)', 'Queenstown (Q)', 'Southampton (S)'])
plt.xticks(rotation=0)
plt.show()
```
